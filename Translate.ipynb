```python
from aksharamukha import transliterate

tamil_text = "இறைவனது திருவடிகளாகிய தெப்பத்தைப் பற்றியவர்களே பிறவியாகிய பெரிய கடலைக்கடப்பர் மற்றவர் கடக்கமாட்டார்"
romanized_text = transliterate.process('Tamil', 'IAST', tamil_text)
print(romanized_text)  # Output: tamiḻ oru aḻakāṉa moḻi
```

```python
# prompt: import the fyp data and  and romanise all the clean_text rows to row called R_clean_text

import pandas as pd
from aksharamukha import transliterate
from unidecode import unidecode

# Assuming your FYP data is in a CSV file named 'fyp_data.csv'
# Replace 'fyp_data.csv' with the actual filename if different
try:
    df = pd.read_csv('fyp_data.csv')
except FileNotFoundError:
    print("Error: 'fyp_data.csv' not found. Please upload the file or provide the correct filename.")
    exit()


def romanize_tamil(text):
    try:
        romanized_text = transliterate.process('Tamil', 'ISO', text)
        text_without_diacritics = unidecode(romanized_text)
        return text_without_diacritics
    except Exception as e:
        print(f"Error romanizing text: {text}. Error: {e}")
        return ""  # Or handle the error as needed


# Apply the romanization function to the 'clean_text' column
df['R_clean_text'] = df['clean_text'].apply(romanize_tamil)

# Print or save the updated DataFrame
print(df.head())
# df.to_csv('romanized_fyp_data.csv', index=False) #Optional: Save the updated dataframe to a new csv file

```
